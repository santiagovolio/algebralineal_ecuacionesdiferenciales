````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar el dataset
df = pd.read_csv('salary_dataset_lab(1).csv')

# 2. Mostrar las primeras 5 filas
print("--- Primeras 5 filas ---")
print(df.head())

# 3. Información general
print("\n--- Información General ---")
print(df.info())

# 4. Resumen estadístico
print("\n--- Resumen Estadístico ---")
print(df.describe())
````

Nuestro análisis muestra que la experiencia es un motor principal del salario en este dataset, con una correlación de 0.8034. El gráfico de dispersión evidencia que, aunque existen pequeñas variaciones, la tendencia es consistente y lineal, lo que hace que la experiencia sea una variable fundamental para que RRHH tome decisiones sobre compensaciones.

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 1. Cargar dataset (nombre correcto del archivo)
df = pd.read_csv("salary_dataset_lab(1).csv")

# 2. Variables independientes y dependiente
X = df[["years_experience", "education_level", "age"]]
y = df["salary"]

# 3.  Crear y entrenar el modelo
model = LinearRegression()
model.fit(X, y)

# Mostrar coeficientes
coeficientes = pd.DataFrame({
    "Variable": ["Intercepto", "Years Experience", "Education Level", "Age"],
    "Coeficiente": [model.intercept_, *model.coef_]
})

print("Coeficientes del modelo:")
print(coeficientes)

# 4.  Y  fianlemnte métrica de ajuste
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print("\nR² del modelo:", round(r2, 3))
````

Nuestro análisis se basó en construir un modelo que toma en cuenta varias variables al mismo tiempo: la experiencia, el nivel educativo y la edad. A diferencia del modelo anterior, este nos permite ver el efecto de cada factor sin analizarlos de forma aislada.

Al evaluar el modelo, el valor de R2 muestra que el ajuste mejora cuando incluimos más variables, lo que indica que el salario se explica mejor de esta manera. Esto nos lleva a concluir que el salario no depende únicamente de la experiencia laboral, sino de una combinación de distintos factores.
